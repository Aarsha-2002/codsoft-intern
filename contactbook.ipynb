```python
class ContactBook:
    def __init__(self):
        self.contacts = []

    def add_contact(self, name, number):
        self.contacts.append((name, number))
        print(f"Added contact: {name} - {number}")

    def update_contact(self, name, new_number):
        for i, (n, _) in enumerate(self.contacts):
            if n == name:
                self.contacts[i] = (name, new_number)
                print(f"Updated contact: {name} - {new_number}")
                return
        print(f"{name} not found in contacts. Use add_contact to add this contact.")

    def delete_contact(self, name):
        for i, (n, _) in enumerate(self.contacts):
            if n == name:
                del self.contacts[i]
                print(f"Deleted contact: {name}")
                return
        print(f"{name} not found in contacts.")

    def search_contact(self, name):
        for n, number in self.contacts:
            if n == name:
                print(f"Contact details for {name}: {number}")
                return
        print(f"{name} not found in contacts.")

    def list_contacts(self):
        if not self.contacts:
            print("Contact book is empty.")
        else:
            print("List of contacts:")
            for name, number in self.contacts:
                print(f"{name} - {number}")

if __name__ == "__main__":
    contact_book = ContactBook()

    contact_book.add_contact("Alice", "1234567890")
    contact_book.add_contact("Bob", "9876543210")

    contact_book.list_contacts()

    contact_book.update_contact("Alice", "9999999999")

    contact_book.search_contact("Bob")

    contact_book.delete_contact("Bob")

    contact_book.list_contacts()
```
