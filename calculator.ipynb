```python
def add(x, y):
    return x + y
def sub(x, y):
    return x - y
def multi(x, y):
    return x * y
def div(x, y):
    return x / y if y != 0 else "Error: Division by zero!"

while True:
    print("Select operation:")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("5. Exit")

    a = input("Enter choice (1/2/3/4/5): ")

    if a == '5':
        break

    if a in ('1', '2', '3', '4'):
        n1 = float(input("Enter first number: "))
        n2 = float(input("Enter second number: "))

        if a == '1':
            print(n1, "+", n2, "=", add(n1, n2))
        elif a == '2':
            print(n1, "-", n2, "=", sub(n1, n2))
        elif a == '3':
            print(n1, "*", n2, "=", multi(n1, n2))
        elif a == '4':
            res = div(n1, n2)
            print(n1, "/", n2, "=", res)
    else:
        print("Invalid input")
```
